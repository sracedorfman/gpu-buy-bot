```python
def wait():
    # time.sleep(random.randrange(2,5))
    time.sleep(0)
```

```python
def check_in_stock(r):
    buy_btn = r.html.find('button[class="btn btn-primary btn-lg btn-block btn-leading-ficon add-to-cart-button"]', first=True)
    return (buy_btn is not None)
```

```python
def added_to_cart(icon):
    return icon.get_attribute("aria-label") != "Cart, 0 item"
```

```python
def perform_purchase(wd, url):
    try:
        wd.refresh()
    except:
        wd = wdr.Chrome()
        wd.implicitly_wait(10)
        wd.get(base_url)
    wd.switch_to.window(wd.current_window_handle)
    # add_to_cart = wd.find_element_by_css_selector('button[class="btn btn-primary btn-lg btn-block btn-leading-ficon add-to-cart-button"]')
    # added_to_cart = False
    cart_total = wd.find_element_by_css_selector('a[data-lid="hdr_carticon"]')
    # added_to_cart = cart_total.get_attribute("aria-label") == "Cart, 1 item"
    # print(added_to_cart)
    add_to_cart = wd.find_element_by_css_selector('button[class*="btn-primary"]')
    while not added_to_cart(cart_total):
        try:
            add_to_cart.click()
        except Exception as e:
            print(e)
#     while not added_to_cart:
#         try:
#             add_to_cart = wd.find_element_by_css_selector('button[class*="btn-primary"]')
#             add_to_cart.click()
#             try:
#                 close = wd.find_element_by_css_selector('button[class="btn-default-link close-modal-x"]')
#                 close.click()
#                 added_to_cart = True
#             except:
#                 print('')
#         except:
#             print('')
    # add_to_cart.click()
    try:
        close = wd.find_element_by_css_selector('button[class="btn-default-link close-modal-x"]')
        close.click()
    except:
        print('pain')
    finally:
        cart = wd.find_element_by_css_selector('span[class="cart-label"]')
        cart.click()
    wait()
    checkout = wd.find_element_by_css_selector('button[class="btn btn-lg btn-block btn-primary"]')
    checkout.click()
    wait()
    
    email = info.email
    password = info.password
    first_name = info.first_name
    last_name = info.last_name
    street_address = info.street_address
    city = info.city
    state = info.state
    zip_code = info.zip_code
    cvv = info.cvv
    
    wd.find_element_by_id('fld-e').send_keys(email)
    wait()
    wd.find_element_by_id('fld-p1').send_keys(password)
    wait()
    login = wd.find_element_by_css_selector('button[class="btn btn-secondary btn-lg btn-block c-button-icon c-button-icon-leading cia-form__controls__submit "]')
    login.click()
    wait()
    
    try:
#        first_name = wd.find_element_by_css_selector('input[id=consolidatedAddresses.ui_address_1134.firstName')
        first_name = wd.find_element_by_css_selector('input[id*=firstName')
        first_name.send_keys(first_name)
        wait()
        wd.find_element_by_css_selector('input[id$=lastName').send_keys(last_name)
        wait()
        wd.find_element_by_css_selector('input[id$=street').send_keys(street_address)
        wait()
        wd.find_element_by_css_selector('input[id$=city').send_keys(city)
        wait()
        select = Select(wd.find_element_by_css_selector('select[id$=state'))
        select.select_by_value(state)
        wait()
        wd.find_element_by_css_selector('input[id$=zipcode').send_keys(zip_code)
    except Exception as e:
        try:
            continue_to_payment = wd.find_element_by_css_selector('button[class="btn btn-lg btn-block btn-secondary"]')
            continue_to_payment.click()
        except Exception as e:
            #try:
            #pickup = wd.find_element_by_css_selector('a[class$="ispu-card__switch"]')
            #pickup.click()
            time.sleep(10)
            start_time = time.time()
            clickable = False
            while not clickable:
                try:
                    select_location = wd.find_element_by_css_selector('div[data-test-id="select-store-button"]')
                    select_location.click()
                    clickable = True
                except:
                    print('not clickable yet')
                    if (time.time() - start_time) > 30:
                        clickable = True
    finally:
        wait()
        continue_to_payment = wd.find_element_by_css_selector('button[class="btn btn-lg btn-block btn-secondary"]')
        continue_to_payment.click()
        wait()
        wd.find_element_by_css_selector('input[id=credit-card-cvv').send_keys(cvv)
        wait()
        place_order = wd.find_element_by_css_selector('button[data-track^="Place your Order"]')
        if place_order is not None:
            print('huzzah!')
        place_order.click()
    
    
    
```

```python
session = HTMLSession()
r = session.get(base_url)
i = 1
wd = wdr.Chrome()
wd.implicitly_wait(10)
wd.get(base_url)
ref_time_sum = 0
while not check_in_stock(r):
    start_time = time.time()
    if i > 1 and (i % (10**(math.floor(math.log10(i))))) == 0 :
        print('%d; %.3f seconds' % (i, ref_time_sum/(i-1)))
    i = i + 1
    time.sleep(random.randrange(10)/100.0)
    r = session.get(base_url)
    ref_time_sum += time.time() - start_time
print('In stock!')
perform_purchase(wd, base_url)
print('purchase_placed')
```
